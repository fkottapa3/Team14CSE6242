```python
import pandas as pd
import numpy as np
import shutil
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report



#READ DATASET

data= pd.read_csv("C:\\Users\\vijay\\Desktop\\Masters - OMS\\CSE 6242\\Vijay\\project\\diabetes_prediction_dataset.csv")
                      

#VIEW DATA SET AND COLUMNS
                      
display(data.info(),data.head())

#gender: Gender of the individual (e.g., Female, Male)
#age: Age of the individual
#hypertension: Whether the individual has hypertension (0 = No, 1 = Yes)
#heart_disease: Whether the individual has heart disease (0 = No, 1 = Yes)
#smoking_history: Smoking history of the individual (e.g., never, current, No Info)
#bmi: Body Mass Index (BMI) of the individual
#HbA1c_level: HbA1c level of the individual (used to diagnose diabetes)
#blood_glucose_level: Blood glucose level of the individual
#diabetes: Target variable indicating whether the individual has diabetes (0 = No, 1 = Yes)


#DATA PROCESSING

#checking for missing values
missing_values = data.isnull().sum()
print(missing_values)


# Check the number of duplicate rows
print(data.duplicated().sum())

# Remove duplicate rows and reassign to 'data'
data = data.drop_duplicates()

# Verify that duplicates have been removed
print("______Removed Duplicate______")
print(data.duplicated().sum())


# List of numerical columns to check for outliers
#columns_to_check = ['bmi', 'HbA1c_level', 'blood_glucose_level']

#removed bmi as it has many outliers
columns_to_check = ['HbA1c_level', 'blood_glucose_level']

# Plot box plots for each numerical column
plt.figure(figsize=(12, 6))
for i, col in enumerate(columns_to_check, 1):
    plt.subplot(1, len(columns_to_check), i)
    sns.boxplot(y=data[col])
    plt.title(f"Box Plot of {col}")

plt.tight_layout()
plt.show()

# Function to remove outliers using IQR
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)  # First quartile
    Q3 = df[column].quantile(0.75)  # Third quartile
    IQR = Q3 - Q1  # Interquartile Range
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Remove outliers
    df_cleaned = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df_cleaned

# Apply outlier removal to all selected columns
data_cleaned = data.copy()
for col in columns_to_check:
    data_cleaned = remove_outliers_iqr(data_cleaned, col)

# Print dataset size before and after removing outliers
print(f"Original dataset size: {data.shape[0]}")
print(f"Dataset size after removing outliers: {data_cleaned.shape[0]}")

# Count the number of rows where 'smoking_history' has 'No Info'
no_info_count = data['smoking_history'].eq('No Info').sum()

print(f'Number of rows with "No Info" in smoking_history: {no_info_count}')

# Replace 'No Info' with NaN in 'smoking_history' column
data['smoking_history'] = data['smoking_history'].replace('No Info', np.nan)

# Impute missing values for 'smoking_history' with the most frequent value
#data['smoking_history'] = data['smoking_history'].fillna(data['smoking_history'].mode()[0])

# Verify the changes
#print(data['smoking_history'].isnull().sum())

data_cleaned.to_csv('modified_diabetes_data.csv', index=False)

#download it within a Jupyter notebook environment
shutil.move('modified_diabetes_data.csv', 'C:\\Users\\vijay\\Desktop\\Masters - OMS\\CSE 6242\\Vijay\\project\\modified_diabetes_data.csv')




# Visualize the distribution of the target variable
plt.figure(figsize=(6, 4))
ax = sns.countplot(x='diabetes', data=data)

# Add value labels on top of each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', 
                (p.get_x() + p.get_width() / 2, p.get_height()), 
                ha='center', va='bottom', fontsize=12, fontweight='bold')

# Title and display
plt.title("Distribution of Diabetes")
plt.xlabel("Diabetes (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()

# Plot distributions of numerical features
numerical_cols = ['age', 'bmi', 'HbA1c_level', 'blood_glucose_level']

# Create subplots for each numerical column
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(2, 2, i)
    ax = sns.histplot(data[col], kde=True, bins=20)

    # Add value labels on bars
    for p in ax.patches:
        if p.get_height() > 0:  # Avoid annotating empty bins
            ax.annotate(f'{int(p.get_height())}', 
                        (p.get_x() + p.get_width() / 2, p.get_height()), 
                        ha='center', va='bottom', fontsize=10, fontweight='bold')

    plt.title(f'Distribution of {col}')

plt.tight_layout()
plt.show()


numeric_data = data.select_dtypes(include=['number'])

# Compute the correlation matrix
corr_matrix = numeric_data.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()


# Pairplot to visualize relationships between numerical columns
sns.pairplot(data, hue='diabetes', vars=numerical_cols)
plt.show()


# Visualize gender distribution
def add_labels(ax):
    for p in ax.patches:
        ax.annotate(f'{int(p.get_height())}', 
                    (p.get_x() + p.get_width() / 2, p.get_height()), 
                    ha='center', va='bottom', fontsize=10, fontweight='bold')

# Visualize gender distribution with diabetes
plt.figure(figsize=(8, 6))
ax = sns.countplot(x='gender', data=data, hue='diabetes')
add_labels(ax)  # Add labels
plt.title("Gender Distribution with Diabetes")
plt.show()

# Visualize smoking history distribution with diabetes
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='smoking_history', data=data, hue='diabetes')
add_labels(ax)  # Add labels
plt.title("Smoking History Distribution with Diabetes")
plt.xticks(rotation=45)
plt.show()

#Analyzing the Relationship Between Features and Target Variable

# Age vs Diabetes
plt.figure(figsize=(8, 6))
sns.boxplot(x='diabetes', y='age', data=data)
plt.title("Age vs Diabetes")
plt.show()


# BMI vs Diabetes
plt.figure(figsize=(8, 6))
sns.boxplot(x='diabetes', y='bmi', data=data)
plt.title("BMI vs Diabetes")
plt.show()

# HbA1c Level vs Diabetes
plt.figure(figsize=(8, 6))
sns.boxplot(x='diabetes', y='HbA1c_level', data=data)
plt.title("HbA1c Level vs Diabetes")
plt.show()

# Blood Glucose Level vs Diabetes
plt.figure(figsize=(8, 6))
sns.boxplot(x='diabetes', y='blood_glucose_level', data=data)
plt.title("Blood Glucose Level vs Diabetes")
plt.show()

# Pairwise relationships with diabetes using scatterplots
sns.pairplot(data, hue='diabetes', markers='o', plot_kws={'alpha': 0.6})
plt.show()

```
